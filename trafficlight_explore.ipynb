# Exploring LISA Traffic Light Dataset

The LISA Traffic Light Dataset is a collection of annotated images that can be used for training computer vision models to recognize traffic lights. The analysis of the LISA Traffic Light Dataset is performed using Python and various data analysis libraries. This notebook shows exploration around the dataset for daySequence1 and nightSequence1. 

Dataset:

daySequence1 = 4060 frames,
nightSequence1 = 4993 frames,

```python
# Import Libraries
import cv2
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# DaySequence1, NightSequence1

```python
# DaySequence1 annotation file path
anno_path_box_day = "Annotations/Annotations/daySequence1/frameAnnotationsBOX.csv"
# NightSequence1 annotation file path
anno_path_box_night = "Annotations/Annotations/nightSequence1/frameAnnotationsBOX.csv"

# Path to image for DaySequence1, random selection of image
frames_path_day = "frames/daySequence1/daySequence1/frames/"
print("Choosing random image from daySeqence1 frame data...")
frame_id_day = np.random.choice(os.listdir(frames_path_day))


frame_path_day = os.path.join(frames_path_day, frame_id_day)
print("Using image:\t'{}'\nLocation:\t'{}'\n".format(frame_id_day, frame_path_day))

# Path to image for NightSequence1, random selection of image
frames_path_night = "frames/nightSequence1/nightSequence1/frames/"
print("Choosing random image from nightSeqence1 frame data...")
frame_id_night = np.random.choice(os.listdir(frames_path_night))

frame_path_night = os.path.join(frames_path_night, frame_id_night)
print("Using image:\t'{}'\nLocation:\t'{}'".format(frame_id_night, frame_path_night))
```

```python
# Read 'anno_path_box_day' file and capture the file needed for drawing bounding box
annotations_day = []
with open(anno_path_box_day) as fp:  
    line = fp.readline()
    line = fp.readline() 
    #cnt = 1
    while line:
        anno_file_path = (line.strip()).split(";")
        anno_file_id = anno_file_path[0].split("/")[1]
        if anno_file_id == frame_id_day:
            annotations_day.append(anno_file_path)
        line = fp.readline()
print("Annotations captured for image ", frame_id_day)
print(annotations_day)
```

```python
# Read 'anno_path_box_night' file and capture the file needed for drawing bounding box
annotations_night = []
with open(anno_path_box_night) as fp:  
    line = fp.readline()
    line = fp.readline() 
    #cnt = 1
    while line:
        anno_file_path = (line.strip()).split(";")
        anno_file_id = anno_file_path[0].split("/")[1]
        if anno_file_id == frame_id_night:
            annotations_night.append(anno_file_path)
        line = fp.readline()
print("Annotations captured for image ", frame_id_night)
print(annotations_night)
```

```python
# Plot annotations on image for daySequence1
img_day = cv2.imread(frame_path_day)
img_day = cv2.cvtColor(img_day, cv2.COLOR_BGR2RGB)

# Original Image without Annotations
print("Original Image ", frame_id_day)
plt.rcParams['figure.figsize'] = [12, 6]
plt.imshow(img_day)
plt.show()
```

```python
# Plot annotations on image for nightSequence1
img_night = cv2.imread(frame_path_night)
img_night = cv2.cvtColor(img_night, cv2.COLOR_BGR2RGB)

# Original Image without Annotations
print("Original Image ", frame_path_night)
plt.rcParams['figure.figsize'] = [12, 6]
plt.imshow(img_night)
plt.show()
```

```python
# Annotated Image for daySequence1 
print("Total annotations {}:".format(len(annotations_day)))

for annotation in annotations_day:
    print(annotation)
    anno_class = annotation[1]
    x1 = int(annotation[2])
    y1 = int(annotation[3])
    x2 = int(annotation[4])
    y2 = int(annotation[5])
    print("\tClass: '{}' at [{},{},{},{}]".format(anno_class, x1, y1, x2, y2))
    if anno_class == "go" or anno_class == "goLeft" or anno_class == "goForward":
        color_class = (0,255,0)
    elif anno_class == "warning" or anno_class == "warningLeft":
        color_class = (255,255,0)
    elif anno_class == "stop" or anno_class == "stopLeft":
        color_class = (255,0,0)
    cv2.rectangle(img_day, (x1, y1), (x2, y2), color_class, 2)
    cv2.putText(img_day, anno_class, (x1, y1-5),
                                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, color_class, 2, 
                                        lineType=cv2.LINE_AA)

plt.imshow(img_day)
plt.show()
```

```python
# Annotated Image for nightSequence1 
print("Total annotations {}:".format(len(annotations_night)))

for annotation in annotations_night:
    print(annotation)
    anno_class = annotation[1]
    x1 = int(annotation[2])
    y1 = int(annotation[3])
    x2 = int(annotation[4])
    y2 = int(annotation[5])
    print("\tClass: '{}' at [{},{},{},{}]".format(anno_class, x1, y1, x2, y2))
    if anno_class == "go" or anno_class == "goLeft" or anno_class == "goForward":
        color_class = (0,255,0)
    elif anno_class == "warning" or anno_class == "warningLeft":
        color_class = (255,255,0)
    elif anno_class == "stop" or anno_class == "stopLeft":
        color_class = (255,0,0)
    cv2.rectangle(img_night, (x1, y1), (x2, y2), color_class, 2)
    cv2.putText(img_night, anno_class, (x1, y1-5),
                                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, color_class, 2, 
                                        lineType=cv2.LINE_AA)

plt.imshow(img_night)
plt.show()
```
